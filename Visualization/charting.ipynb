```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use("dark_background")
```

```python
df = pd.read_csv("trending.csv", index_col="Unnamed: 0")
```

```python
df.columns
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,5))

axes[0].scatter(df["popularity"], df["vote_average"], alpha=0.5, color="teal")
axes[0].set_title("Popularity vs Average Votes")
axes[0].set_xlabel("Popularity Score")
axes[0].set_ylabel("Average Vote (0-10)")
axes[0].grid(True, alpha=0.3)

axes[1].hist(df["vote_average"], bins=20, color="skyblue", edgecolor="black")
axes[1].set_title("Distribution of Movie Ratings")
axes[1].set_xlabel("Vote Average")
axes[1].set_ylabel("Frequency")
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
