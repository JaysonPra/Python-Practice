```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use("default")
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head(5)
```

```python
tips.columns.to_list()
```

```python
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker"
)
```

```python
sns.barplot(
    data=tips,
    x="sex",
    y="size",
    estimator=np.mean
)
```

```python
sns.barplot(
    data=tips,
    x="smoker",
    y="tip",
    estimator=np.mean
)
```
